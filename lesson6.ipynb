Импорт библиотек

```python
from keras_cv.models import StableDiffusionV2
import matplotlib.pyplot as plt
```

Создаем модель для генерации изображений

```python
stable_diffusion = StableDiffusionV2(512, 512)
```

Генерируем изображения

```python
images = stable_diffusion.text_to_image(
    prompt="A zoomed out photograph of a cool looking cat. The cat stands in a beautiful forest",
    negative_prompt="unrealistic, bad looking, malformed",
    batch_size=4,
    seed=1231,
)
```

Отобразим сгенерированные изображения

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for img, ax in zip(images, axs.flatten()):
    ax.imshow(img)
    ax.axis('off')
plt.show()
```
